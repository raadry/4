```python
#Assignment 3
#Raad Khan


print("Express for 5 items or fewer. Regular otherwise. \n")
print ("Please enter select either Express or Regular checkout")
select = input("Enter Express or Regular:")
while select not in ("Express", "Regular", "regular", "express"):
    print ("Invalid response.")
    select = input("Try again:")
    
count = 0

if select == "Express" or select == "express":
    itemlist = ""
    
    
    moreitems='Y'
    finalprice = 0
    while moreitems == 'Y' or moreitems == 'y':
        count=count+1
        print("Item number ",count)
        if count > 5:
            break
        else:
            item = input("Please name the item: ")
            quant = int(input("How many items? "))
            price = float(input("Please enter the price: "))
            totalprice =  (price*quant)
            finalprice = finalprice+totalprice    
            itemlist += item + "\t" + str(totalprice) + "\n"
            
            moreitems = input("Are there more items? (Y/N): ")
            while moreitems not in ("Y","y","N","n"):
                print("Invalid response.")
                moreitems = input("Try again: ")
            

    
    
            
    
    print ("Item \t Price")
    print ("-----------------------")
    print(itemlist)
    print ("\nTotal \t", finalprice)
    
elif select == "Regular" or select == "regular":
    itemlist = ""
    
    
    moreitems='Y'
    finalprice = 0
    while moreitems == 'Y' or moreitems == 'y':
        count=count+1
        print("Item number ",count)
        

        item = input("Please name the item: ")
        quant = int(input("How many items? "))
        price = float(input("Please enter the price: "))
        totalprice =  (price*quant)
        finalprice = finalprice+totalprice    
        itemlist += item + "\t" + str(totalprice) + "\n"
        moreitems = input("Are there more items? (Y/N): ")
        while moreitems not in ("Y","y","N","n"):
            print("Invalid response.")
            moreitems = input("Try again: ")
        
    
    print ("Item \t Price")
    print ("-----------------------")
    print(itemlist)
    print ("\nTotal \t", finalprice)

print ("\n\nHow would you like to pay?")
print ("Enter 1 for Cash")
print ("Enter 2 for Credit Card")

choice1 = int(input("Please enter your response: "))

if choice1 == 1:
    cash = float(input("Please enter your cash value inserted: "))
    if cash > finalprice:
        print ("You entered $", "{:.2f}".format(cash))
        print ("Your change is $", "{:.2f}".format(cash-finalprice))
        print ("Thank you for your purchase.")
    elif cash == finalprice:
        print("Thank you for your purchase.")
    elif cash < finalprice:
        print("You have not paid the full amount. Your items will be returned to the shelves. Thank you!")
    else:
        print("Invalid response.")
elif choice1 == 2:
    print ("$","{:.2f}".format(finalprice), " was charged from your credit card.")
    print ("Thank you for your purchase.")
else :
    print ("Invalid entry.")
```
